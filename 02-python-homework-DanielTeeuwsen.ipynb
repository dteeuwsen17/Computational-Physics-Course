```python

    if expression(i):
        new_list.append(old_list[i])
```

**B.** Second, do it by *list comprehensions*, for example:
```python

```

**C.** Third, simplify the list comprehension, for example: 


Make certain that your the varaible you `print` to see the output is a NumPy array.

````python
##Part A

old_list=np.arange(14) #technically an array not a list
new_list=[]            #actually a list
for i in old_list:
    if i%2==1:
        new_list.append(old_list[i])
new_list=np.array(new_list)
print(new_list)
````

````python
##Part B

old_listB=np.arange(14)
new_listB=np.array([old_listB[i] for i in old_listB if old_listB[i]%2==1])
print(new_listB)
````

````python
##Part C

old_listC=np.arange(14)
new_listC=old_listC[old_listC%2==1]
print(new_listC)
````
